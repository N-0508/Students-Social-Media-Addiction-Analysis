# Students Social Media Addiction Analysis

### ABOUT THE PROJECT 

This project focuses on analyzing student social media usage patterns using Exploratory Data Analysis (EDA).  
The dataset includes information about screen time, platform usage, study hours, and daily habits of students.  

The main objective of this analysis is to understand how social media affects students' productivity, sleep, and overall lifestyle.  
By using data visualization and statistical analysis, we aim to identify key patterns, trends, and relationships in the data.  

This project helps in gaining insights into whether increased social media usage leads to changes in academic performance and daily routines.

```python
# importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# loading the datasets 
df = pd.read_csv("Students Social Media Addiction.csv")
```

### UNIVARIATE ANALYSIS

```python
# Age Distribution
plt.figure()
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Daily Usage Hours
plt.figure()
sns.histplot(df['Avg_Daily_Usage_Hours'], kde=True)
plt.title("Daily Social Media Usage")
plt.show()
```

```python
# Sleep Hours
plt.figure()
sns.histplot(df['Sleep_Hours_Per_Night'], kde=True)
plt.title("Sleep Hours Distribution")
plt.show()
```

```python
# Addiction Score
plt.figure()
sns.histplot(df['Addicted_Score'], kde=True)
plt.title("Addiction Score Distribution")
plt.show()
```

### CATEGORICAL ANALYSIS

```python
# Gender Count
plt.figure()
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# Academic Level
plt.figure()
sns.countplot(x='Academic_Level', data=df)
plt.title("Academic Level Distribution")
plt.show()
```

```python
# Most Used Platform
plt.figure()
sns.countplot(x='Most_Used_Platform', data=df)
plt.title("Most Used Social Media Platform")
plt.xticks(rotation=45)
plt.show()
```

```python
# Academic Performance Impact
plt.figure()
sns.countplot(x='Affects_Academic_Performance', data=df)
plt.title("Impact on Academic Performance")
plt.show()
```

### BIVARIATE ANALYSIS

```python
# Usage vs Sleep
# Create bins
df['Usage_Category'] = pd.cut(df['Avg_Daily_Usage_Hours'],
                             bins=[0, 2, 4, 6, 8, 10],
                             labels=['0-2', '2-4', '4-6', '6-8', '8-10'])

# Plot
plt.figure(figsize=(8,5))
sns.boxplot(x='Usage_Category', y='Mental_Health_Score', data=df)
plt.title("Usage vs Mental Health")
plt.show()
```

```python
# Usage vs Mental Health
plt.figure(figsize=(8,5))
sns.scatterplot(x='Avg_Daily_Usage_Hours', y='Mental_Health_Score', data=df)
plt.title("Usage vs Mental Health")
plt.show()
```

```python
# Usage vs Addiction Score
plt.figure()
sns.scatterplot(x='Avg_Daily_Usage_Hours', y='Addicted_Score', data=df)
plt.title("Usage vs Addiction Score")
plt.show()
```

```python
# Conflicts vs Addiction
plt.figure()
sns.boxplot(x='Conflicts_Over_Social_Media', y='Addicted_Score', data=df)
plt.title("Conflicts vs Addiction")
plt.show()
```

### CORRELATION ANALYSIS

```python
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

### MULTIVARIATE ANALYSIS

```python
# Pairplot (important)
sns.pairplot(df[['Avg_Daily_Usage_Hours',
                 'Sleep_Hours_Per_Night',
                 'Mental_Health_Score',
                 'Addicted_Score']])
plt.show()
```

### GROUPED INSIGHTS

```python
# Average addiction by gender
print("\nAddiction Score by Gender:")
print(df.groupby('Gender')['Addicted_Score'].mean())
```

```python
# Addiction by academic level
print("\nAddiction Score by Academic Level:")
print(df.groupby('Academic_Level')['Addicted_Score'].mean())
```

```python
# Platform vs addiction
print("\nAddiction Score by Platform:")
print(df.groupby('Most_Used_Platform')['Addicted_Score'].mean())
```
